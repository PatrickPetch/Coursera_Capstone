# Beginning of Part 3

Plot the map of data in each row

```python
latitude = 43.71
longitude  = -79.38
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

for lat, lon, nei, bor, code in zip(result_df['Latitude'], result_df['Longitude'], result_df['Neighborhood'], result_df['Borough'], result_df['Postal Code']):
    label = folium.Popup(code)
    folium.CircleMarker(
        [lat, lon],
        radius=4,
        popup=label,
        color='coral',
        fill=True,
        fill_color='white',
        fill_opacity=0.7
    ).add_to(map_clusters)
       
map_clusters
```

Let's separate the data points into 5 clusters and label them with different colour according to cluster.

Create an array which tells which cluster each data point belongs to.

```python
X = np.array([[x, y] for x, y in zip(result_df['Latitude'], result_df['Longitude'])])
kmeans = KMeans(init='k-means++', n_clusters=5, random_state=0).fit(X)
x = kmeans.labels_
x = list(x)
print(x)
```

Plot the same map but change the colour of each data point according to the cluster of each point.

```python
latitude = 43.71
longitude  = -79.38
map_clusters1 = folium.Map(location=[latitude, longitude], zoom_start=11)

colour = ['blue', 'green', 'coral', 'yellow', 'purple', 'pink']
for lat, lon, nei, bor, code, group in zip(result_df['Latitude'], result_df['Longitude'], result_df['Neighborhood'], result_df['Borough'], result_df['Postal Code'], x):
    label = folium.Popup(code)
    folium.CircleMarker(
        [lat, lon],
        radius=4,
        popup=label,
        color=colour[group],
        fill=True,
        fill_color='white',
        fill_opacity=0.7
    ).add_to(map_clusters1)
       
map_clusters1
```
